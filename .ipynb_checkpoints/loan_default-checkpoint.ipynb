```python
import pandas as pd
```

```python
# load train and test data
train_data = pd.read_csv("credit-default-prediction-ai-big-data/train.csv")
test_data = pd.read_csv("credit-default-prediction-ai-big-data/test.csv")
```

```python
# shape of the train_data and test data
train_data.shape,test_data.shape
```

```python
# head value
train_data.head(10)
```

```python
# remove id and Credit Default
y_target =train_data["Credit Default"].values
train_data = train_data.drop(["Id", "Credit Default"], axis=1)
test_data = test_data.drop("Id", axis=1)
```

```python
# size of train and test data
train_size = train_data.shape[0]
test_size = test_data.shape[0]
```

```python
# combine train and test data to full_data
full_data = pd.concat([train_data, test_data], axis=0)
```

```python
full_data.head()
```

```python
# check columns NaN or not
full_data.isnull().sum()
```

```python
# columns 
all_col_names = list(full_data.columns)
num_col_names = list(full_data._get_numeric_data().columns)
cat_col_names = list(set(all_col_names) - set(num_col_names))
```

```python
len(all_col_names),len(num_col_names),len(cat_col_names)
```

```python
# fill NONe in categorical columns
for i in cat_col_names:
    full_data[i] = full_data[i].fillna("NONE")
```

```python
full_data.head()
```

```python
# fill null values with mean values
for i in num_col_names:
    mean_val = full_data[i].mean()
    full_data[i] = full_data[i].fillna(mean_val)
```

```python
full_data.head()
```

```python
full_data.isnull().sum()
```

```python
# label encoding 
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for i in cat_col_names:
    label_encoder.fit(full_data[i])
    full_data[i] = label_encoder.transform(full_data[i])
```

```python
full_data.head()
```

```python
# apply normalization to numerical columns
from sklearn.preprocessing import MinMaxScaler
import numpy as np
min_max_scaler = MinMaxScaler()
for i in num_col_names:
    full_data[i] = min_max_scaler.fit_transform(np.array(full_data[i]).reshape(-1,1))
```

```python
full_data.head()
```

```python
x_train = full_data[:train_size]
x_test = full_data[train_size:]
```

```python
x_train.shape,x_test.shape, y_target.shape
```

```python
# build linear svm
from sklearn.svm import LinearSVC
lsvc_clf = LinearSVC(tol=1e-5,
                     verbose=2,
                    random_state=0,
                     max_iter=2000
                    )
# fit model
lsvc_clf.fit(x_train, y_target)
# predictions using model
predicts = lsvc_clf.predict(x_test)

# create dictionary
pred_dict = {"Credit Default":list(predicts)}

# sample submission data
sample_df = pd.read_csv("credit-default-prediction-ai-big-data/sampleSubmission.csv")

#update submit dataframe with pred_dict
sample_df.update(pred_dict)
#save sample_df as csv file
sample_df.to_csv("submission_1.csv", index=False)
```

```python
# build nearest neighbors
from sklearn.neighbors import KNeighborsClassifier
knc_clf = KNeighborsClassifier(algorithm='auto')
# fit model
knc_clf.fit(x_train, y_target)
# predictions using model
predicts = knc_clf.predict(x_test)

# create dictionary
pred_dict = {"Credit Default":list(predicts)}

# sample submission data
sample_df = pd.read_csv("credit-default-prediction-ai-big-data/sampleSubmission.csv")

#update submit dataframe with pred_dict
sample_df.update(pred_dict)
#save sample_df as csv file
sample_df.to_csv("submission_knn.csv", index=False)
```

```python
# build a model with random forest class
# build nearest neighbors
from sklearn.ensemble import RandomForestClassifier
rfc_clf = RandomForestClassifier(criterion='entropy',
                                verbose=2,
                                random_state=0,
                                n_estimators=200
                                )
# fit model
rfc_clf.fit(x_train, y_target)
# predictions using model
predicts = rfc_clf.predict(x_test)

# create dictionary
pred_dict = {"Credit Default":list(predicts)}

# sample submission data
sample_df = pd.read_csv("credit-default-prediction-ai-big-data/sampleSubmission.csv")

#update submit dataframe with pred_dict
sample_df.update(pred_dict)
#save sample_df as csv file
sample_df.to_csv("submission_rfc.csv", index=False)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# craete model 
model_dict = {"model1":
              {
                  "model_name":RandomForestClassifier(), 
                  "gridsearch":
                  {
                      "n_estimators":[100,200,300,400],
                      "max_depth":[1,2,3,4,5,10],
                      "criterion":["gini","entropy"]
                  }
              },
              "model2":
              {
                  "model_name":ExtraTreesClassifier(),
                  "gridsearch":
                  {
                      "n_estimators":[100,200,300,400],
                      "max_depth":[1,2,3,4,5,10],
                      "criterion":["gini","entropy"]
                  }
              },
#               "model3":
#               {
#                   "model_name":DecisionTreeClassifier(),
#                   "gridsearch":
#                   {
#                       "n_estimators":[100,200,300,400],
#                       "max_depth":[1,2,3,4,5,10],
#                       "criterion":["gini","entropy"]
#                   }
#               }
}
```

```python
from sklearn import model_selection
```

```python
# apply grid search
for each_model in model_dict:
#     print(model_dict[each_model])
    # classifier
    clf = model_dict[each_model]["model_name"]
    # grid parameters
    param_grid = model_dict[each_model]["gridsearch"]
    
    model = model_selection.GridSearchCV(
    estimator=clf,
    param_grid=param_grid,
    scoring='accuracy'
    )
    
    # fit model
    model.fit(x_train,y_target)
    print(f"Best score: {model.best_score_}")
    
    print("Best parameters set:")
    best_parameters = model.best_estimator_.get_params()
    for param_name in sorted(param_grid.keys()):
        print(f"\t{param_name}: {best_parameters[param_name]}") 
```
